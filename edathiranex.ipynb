```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)



url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print("\n First 5 Rows:")
print(df.head())

print("\n Dataset Information:")
print(df.info())

print("\n Statistical Summary:")
print(df.describe())

print("\n Missing Values:")
print(df.isnull().sum())



df['Age'].fillna(df['Age'].median(), inplace=True)

df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)


df.drop('Cabin', axis=1, inplace=True)


df.drop_duplicates(inplace=True)

print("\n Missing Values After Cleaning:")
print(df.isnull().sum())


plt.figure(figsize=(6,4))
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()


plt.figure(figsize=(6,4))
sns.countplot(x='Sex', data=df)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()



plt.figure(figsize=(6,4))
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title('Survival Based on Gender')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()


plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()



plt.figure(figsize=(6,4))
sns.countplot(x='Pclass', data=df)
plt.title('Passenger Class Distribution')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(8,5))
sns.histplot(df['Fare'], bins=30, kde=True)
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()



plt.figure(figsize=(8,5))
sns.boxplot(x=df['Age'])
plt.title('Boxplot of Age')
plt.show()

# ==========================================
# Correlation Matrix
# ==========================================

numeric_df = df.select_dtypes(include=np.number)

corr = numeric_df.corr()

plt.figure(figsize=(12,8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# ==========================================
# Scatter Plot: Age vs Fare
# ==========================================

plt.figure(figsize=(8,5))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()

# ==========================================
# Survival Rate by Passenger Class
# ==========================================

survival_by_class = df.groupby('Pclass')['Survived'].mean()

print("\n Survival Rate by Passenger Class:")
print(survival_by_class)

# ==========================================
# Survival Rate by Gender
# ==========================================

survival_by_gender = df.groupby('Sex')['Survived'].mean()

print("\n Survival Rate by Gender:")
print(survival_by_gender)

# ==========================================
# Key Insights
# ==========================================

print("\n========== KEY INSIGHTS ==========")

print("1. Female passengers had significantly higher survival rates.")

print("2. First-class passengers were more likely to survive.")

print("3. Higher ticket fares were associated with better survival chances.")

print("4. Most passengers were between 20 and 40 years old.")

print("5. Passenger class and gender strongly influenced survival.")

print("6. Some outliers exist in Fare and Age distributions.")

print("==================================")

# ==========================================
# End of Project
# ==========================================
```
